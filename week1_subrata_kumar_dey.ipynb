### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")
    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]
    lhs = A @ vec
    rhs = lam * vec
    assert np.allclose(lhs, rhs), f"Eigenpair {i} failed the Av = λv check!"
    print(f"  eigenpair {i}: λ={lam:.4f} -> A@v matches λ*v ✓")

# Plot both eigenvectors, scaled by their eigenvalues
fig, ax = plt.subplots()
colors = ['crimson', 'darkorange']
for i in range(len(eigenvalues)):
    lam = eigenvalues[i].real
    vec = eigenvectors[:, i].real
    scaled = vec * lam
    ax.quiver(0, 0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1,
              color=colors[i], width=0.01, label=f'λ={lam:.2f}')
ax.set_xlim(-6, 6); ax.set_ylim(-6, 6)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.axhline(0, color='k', lw=0.8); ax.axvline(0, color='k', lw=0.8)
ax.legend(); ax.set_title('Eigenvectors of A, scaled by their eigenvalues')
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

> An eigenvector of a matrix is a direction that the matrix doesn't rotate — when the matrix acts on it, the result still points along the same line (or its exact opposite). The eigenvalue is just the scale factor along that direction: it tells you how much the vector gets stretched (|λ| > 1), shrunk (|λ| < 1), or flipped (λ < 0) as it passes through the transformation. Most vectors get both rotated *and* scaled by a matrix, but eigenvectors are the special directions where only the scaling happens. For our matrix A, every other vector in the plane can be thought of as a mix of the two eigenvector directions, and the matrix stretches each of those two components independently by λ₁=5 and λ₂=2 — which is exactly why eigendecomposition is such a useful way to "see" what a matrix does geometrically.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approximation: keep only the largest singular value/vector pair
S_rank1 = np.zeros_like(S)
S_rank1[0] = S[0]
X_approx = U @ np.diag(S_rank1) @ Vt

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> In `X = U S Vᵀ`, the columns of **V** (rows of Vᵀ) are the **principal component directions** — the axes of maximum variance in the data, ordered from most to least important. The values in **S** (the singular values) tell you how much variance each of those directions explains: squaring and normalizing them gives the same proportions you'd get from the eigenvalues of the covariance matrix in classical PCA. **U** holds the data points' coordinates *in* that new rotated basis, scaled down to unit length. So PCA on a centered data matrix is really just SVD in disguise — instead of computing the covariance matrix and eigendecomposing it, you can run SVD directly on the centered data and read the principal components straight off of `V`, which is also more numerically stable.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    eps = 1e-10

    # Build bin edges from the expected (reference/training) distribution,
    # using percentiles so each bin starts with roughly equal weight.
    breakpoints = np.percentile(expected, np.linspace(0, 100, bins + 1))
    breakpoints[0] = -np.inf
    breakpoints[-1] = np.inf

    expected_counts = np.histogram(expected, bins=breakpoints)[0]
    actual_counts = np.histogram(actual, bins=breakpoints)[0]

    expected_pct = expected_counts / len(expected) + eps
    actual_pct = actual_counts / len(actual) + eps

    psi = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, alpha=0.5, label='Training', color='steelblue', density=True)
ax.hist(drift_dist, bins=30, alpha=0.5, label='Production (drifted)', color='crimson', density=True)
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** The relationship between the input features and the target changes over time — the same input now maps to a different "correct" output (e.g., what counted as fraud last year doesn't match fraud patterns today).
- **Covariate drift:** The distribution of the input features themselves shifts, even though the underlying relationship between inputs and target stays the same (e.g., your user base skews younger than it did during training).
- **PSI < 0.1:** No significant shift — the distributions are close enough that the model's input assumptions still roughly hold.
- **PSI 0.1–0.2:** Minor shift — worth monitoring, may warrant investigation but isn't necessarily urgent.
- **PSI > 0.2:** Major shift — the production distribution has moved substantially from training; the model's performance is at real risk of degrading.
- **Retraining trigger example:** If the PSI on a key feature stays above 0.2 for several consecutive monitoring windows (not just a single noisy spike), that's a reasonable automated trigger to kick off a retraining pipeline using more recent data.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red / total
p_blue  = blue / total
p_green = green / total

# Joint: P(red first AND blue second) — without replacement
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional: P(blue second | red first)
p_blue_given_red = blue / (total - 1)

# Independence check: are P(B|A) == P(B) ?
independent = abs(p_blue_given_red - p_blue) < 1e-9

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.3f}")
print(f"P(blue|red): {p_blue_given_red:.3f}")
print(f"Independent? {independent}")
print("6.1 passed ✓")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = p_free_given_spam * p_spam + p_free_given_ham * p_ham

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam
    p_word = p_word_given_spam * prior_spam + p_word_given_ham * prior_ham
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | 0.30 — our belief that any random email is spam before reading it |
| Likelihood | P(word \| Spam) | 0.80 — how often the word "free" shows up in emails we already know are spam |
| Evidence | P(word) | 0.275 — the overall probability of seeing "free" in any email, spam or not |
| Posterior | P(Spam \| word) | 0.8727 — our updated belief that an email is spam *after* observing that it contains "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# Draw 5,000 samples of size 30 (with replacement) and take their means.
# Vectorized: one (5000, 30) draw instead of looping 5,000 times.
samples = np.random.choice(population, size=(n_samples, sample_size), replace=True)
sample_means = samples.mean(axis=1)

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)   # predicted std of the sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=50, density=True, alpha=0.6, color='steelblue', label='Sample means')
x_range = np.linspace(sample_means.min(), sample_means.max(), 200)
ax.plot(x_range, stats.norm.pdf(x_range, pop_mean, clt_std), color='darkorange', lw=2,
        label='Theoretical N(μ, σ/√n)')
ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> The Central Limit Theorem matters because it lets us treat the *sampling distribution of the mean* as approximately normal regardless of what the original population looks like — here the underlying population is exponential (heavily right-skewed), yet the distribution of sample means still clusters tightly around a bell curve. This is the assumption that makes classical tools like t-tests, z-tests, and confidence intervals valid even when we know real-world data is rarely perfectly normal: as long as the sample size is reasonably large, we can still make principled probabilistic statements about the mean. Note that with n=30 on a strongly skewed population, the match to a true normal isn't perfect (the strict KS test still detects some residual skew) — a useful reminder that "approximately normal" is a matter of degree that improves as n grows, not an instant guarantee.
